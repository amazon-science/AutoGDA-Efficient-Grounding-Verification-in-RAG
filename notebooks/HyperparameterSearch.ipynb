## Naming of Studies
Unfortunately, there is no entirely systematic naming of the hyperparameter studies.
We will list the studies used in the main paper for the datasets in the order (ragtruth-Summ, ragtruth-QA, lfqa, summedits):

* Teacher model best (Vectara, or AlignScore): ```rtsumm_tasksource_final2, rtqa_tasksource_final2, lfqa_alignscore, summall_alignscore```
* Teacher model gpt4 (Vectara, or AlignScore): ```rtsumm_gptq, rtqa_gptq, lfqa_gpt, summall_gptq```
* Teacher model tasksource (self-supervised): ```rtsumm_tasksource_self2, rtqa_tasksource_self2, lfqa_tasksource_self2, summall_tasksource```

````python
from optuna.trial import TrialState
trials_to_keep = []
for t in trial_list:
    if t.state == TrialState.COMPLETE:
        if t.value > 0.5:
            new_trial = optuna.trial.create_trial(
                state=TrialState.COMPLETE,
                params=t.params,
                user_attrs=t.user_attrs,
                system_attrs=t.system_attrs,
                intermediate_values=t.intermediate_values,
                distributions=t.distributions,
                value=t.value
            )
            trials_to_keep.append(t)
newstudy.add_trials(trials_to_keep)
````
